# Sentiment Analysis Project


## Load the Data

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

file_path = '/content/drive/MyDrive/moviereviews.tsv'
df = pd.read_csv(file_path, sep='\t')
df.head()

```

## Remove Blank Records (optional)

```python
df = df[df['review'].notnull() & (df['review'].str.strip() != '')]
df.reset_index(drop=True, inplace=True)
df.head()
```

## Import `SentimentIntensityAnalyzer` and create an sid object
This assumes that the VADER lexicon has been downloaded.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()
```

## Use sid to append a `comp_score` to the dataset

```python
df['comp_score'] = df['review'].apply(lambda x: sid.polarity_scores(x)['compound'])
df[['review', 'comp_score']].head()
```

## Perform a comparison analysis between the original `label` and `comp_score`

```python
from sklearn.metrics import classification_report

df['label'] = df['label'].map({'pos': 'positive', 'neg': 'negative'})

def score_to_sentiment(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['predicted'] = df['comp_score'].apply(score_to_sentiment)
print(classification_report(df['label'], df['predicted']))
```
